```python

#BOOBLE SORT
import random
import matplotlib.pyplot as plt
import time
import numpy as np


def bubble_sort(lista_ordenada):
  n = len(lista_ordenada)
  aux = 0
  for i in range(n-1):
    for j in range(0,n-i-1):
      if lista_ordenada[j] > lista_ordenada[j+1]:
        aux = lista_ordenada[j]
        lista_ordenada[j] =  lista_ordenada[j+1]
        lista_ordenada[j+1] = aux
  return lista_ordenada


def calcular_tempo(lista_ordenada):
  start_time = time.time()
  bubble_sort(lista_ordenada)
  end_time = time.time()
  total_time = end_time - start_time
  return total_time

# Listas para armazenar os tempos
tempos_lista1k = []
tempos_lista10k = []
tempos_lista20k = []
tempo_lista30k = []
tempo_lista40k = []
tempo_lista50k = []
tempos_media = []
tempos_pior_caso = []

tamanhos = [1000, 10000 , 20000, 30000, 40000, 50000]
lista_a_1k = np.random.randint(1, 101, 1000)
lista_b_1k = np.random.randint(1, 101, 1000)
lista_c_1k = np.random.randint(1, 101, 1000)
lista_a_10k = np.random.randint(1, 1001, 10000)
lista_b_10k = np.random.randint(1, 1001, 10000)
lista_c_10k = np.random.randint(1, 1001, 10000)
lista_a_20k = np.random.randint(1, 2001, 20000)
lista_b_20k = np.random.randint(1, 2001, 20000)
lista_c_20k  = np.random.randint(1, 2001, 20000)
lista_a_30k = np.random.randint(1, 3001, 30000)
lista_b_30k = np.random.randint(1, 3001, 30000)
lista_c_30k = np.random.randint(1, 3001, 30000)
lista_a_40k = np.random.randint(1, 4001, 40000)
lista_b_40k = np.random.randint(1, 4001, 40000)
lista_c_40k = np.random.randint(1, 4001, 40000)
lista_a_50k = np.random.randint(1, 5001, 50000)
lista_b_50k = np.random.randint(1, 5001, 50000)
lista_c_50k = np.random.randint(1, 5001, 50000)

def tempo_listas(lista_a, lista_b, lista_c):
  t1 = calcular_tempo(lista_a)
  t2 = calcular_tempo(lista_b)
  t3 = calcular_tempo(lista_c)
  return t1, t2, t3

lista_1k  = list(tempo_listas(lista_a_1k, lista_b_1k, lista_c_1k))
lista_10k = list(tempo_listas(lista_a_10k, lista_b_10k, lista_c_10k))
lista_20k = list(tempo_listas(lista_a_20k, lista_b_20k, lista_c_20k))
lista_30k = list(tempo_listas(lista_a_30k, lista_b_30k, lista_c_30k))
lista_40k = list(tempo_listas(lista_a_40k, lista_b_40k, lista_c_40k))
lista_50k = list(tempo_listas(lista_a_50k, lista_b_50k, lista_c_50k))



media_1k = np.mean(lista_1k)
media_10k = np.mean(lista_10k)
media_20k = np.mean(lista_20k)
media_30k = np.mean(lista_30k)
media_40k = np.mean(lista_40k)
media_50k = np.mean(lista_50k)

lista_media = []
lista_media.append(media_1k)
lista_media.append(media_10k)
lista_media.append(media_20k)
lista_media.append(media_30k)
lista_media.append(media_40k)
lista_media.append(media_50k)


#----------------------MUDAR ESSE METODO DE BUSCAR O WORSTCASE-----------------------
arr_wc_1k = np.arange(1000, 0, -1)
arr_wc_10k = np.arange(10000, 0, -1)
arr_wc_20k = np.arange(20000, 0, -1)
arr_wc_30k = np.arange(30000, 0, -1)
arr_wc_40k = np.arange(40000, 0, -1)
arr_wc_50k = np.arange(50000, 0, -1)


pior_caso_tempos = []

pior_caso_tempos.append(calcular_tempo(arr_wc_1k))
pior_caso_tempos.append(calcular_tempo(arr_wc_10k))
pior_caso_tempos.append(calcular_tempo(arr_wc_20k))
pior_caso_tempos.append(calcular_tempo(arr_wc_30k))
pior_caso_tempos.append(calcular_tempo(arr_wc_40k))
pior_caso_tempos.append(calcular_tempo(arr_wc_50k))

lista_a = [lista_1k[0], lista_10k[0], lista_20k[0],lista_30k[0], lista_40k[0], lista_50k[0]]
lista_b = [lista_1k[1], lista_10k[1], lista_20k[1],lista_30k[1], lista_40k[1], lista_50k[1]]
lista_c = [lista_1k[2], lista_10k[2], lista_20k[2],lista_30k[2], lista_40k[2], lista_50k[2]]


fig , ax = plt.subplots(figsize=(10, 6), layout = 'constrained')
ax.plot(tamanhos,lista_media,
    label='Médias dos tempos',
    color='blue',
    linewidth = 5,
    linestyle = ':')

ax.plot(tamanhos, pior_caso_tempos,
    label='Piores tempos possíveis',
    color='red',
    linestyle = '--',
    linewidth = 2)

ax.plot(tamanhos, lista_a,
    label='Lista A',
    color='green',
    linewidth = 1)

ax.plot(tamanhos, lista_b,
    label='Lista B',
    color='yellow',
    linewidth = 1)

ax.plot(tamanhos, lista_c,
    label='Lista C',
    color='orange',
    linewidth = 1)

ax.set_xlabel('Tamanho da Lista')
ax.set_ylabel('Tempo (segundos)')
ax.set_title('Tempo de Ordenação vs Tamanho da Lista')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import time
import numpy as np

def insert_sort(lista):
  for i in range(len(lista)):
    aux = lista[i]
    j = i - 1
    while j >= 0 and aux < lista[j]:
      lista[j+1] = lista[j]
      j = j - 1
    lista[j+1] = aux
  return lista




def calcular_tempo(lista_ordenada):
  start_time = time.time()
  insert_sort(lista_ordenada)
  end_time = time.time()
  total_time = end_time - start_time
  return total_time



# Listas para armazenar os tempos
tempos_lista1k = []
tempos_lista10k = []
tempos_lista20k = []
tempo_lista30k = []
tempo_lista40k = []
tempo_lista50k = []
tempos_media = []
tempos_pior_caso = []


tamanhos = [1000, 10000 , 20000, 30000, 40000, 50000]
lista_a_1k = np.random.randint(1, 101, 1000)
lista_b_1k = np.random.randint(1, 101, 1000)
lista_c_1k = np.random.randint(1, 101, 1000)
lista_a_10k = np.random.randint(1, 1001, 10000)
lista_b_10k = np.random.randint(1, 1001, 10000)
lista_c_10k = np.random.randint(1, 1001, 10000)
lista_a_20k = np.random.randint(1, 2001, 20000)
lista_b_20k = np.random.randint(1, 2001, 20000)
lista_c_20k  = np.random.randint(1, 2001, 20000)
lista_a_30k = np.random.randint(1, 3001, 30000)
lista_b_30k = np.random.randint(1, 3001, 30000)
lista_c_30k = np.random.randint(1, 3001, 30000)
lista_a_40k = np.random.randint(1, 4001, 40000)
lista_b_40k = np.random.randint(1, 4001, 40000)
lista_c_40k = np.random.randint(1, 4001, 40000)
lista_a_50k = np.random.randint(1, 5001, 50000)
lista_b_50k = np.random.randint(1, 5001, 50000)
lista_c_50k = np.random.randint(1, 5001, 50000)

def tempo_listas(lista_a, lista_b, lista_c):
  t1 = calcular_tempo(lista_a)
  t2 = calcular_tempo(lista_b)
  t3 = calcular_tempo(lista_c)
  return t1, t2, t3

lista_1k  = list(tempo_listas(lista_a_1k, lista_b_1k, lista_c_1k))
lista_10k = list(tempo_listas(lista_a_10k, lista_b_10k, lista_c_10k))
lista_20k = list(tempo_listas(lista_a_20k, lista_b_20k, lista_c_20k))
lista_30k = list(tempo_listas(lista_a_30k, lista_b_30k, lista_c_30k))
lista_40k = list(tempo_listas(lista_a_40k, lista_b_40k, lista_c_40k))
lista_50k = list(tempo_listas(lista_a_50k, lista_b_50k, lista_c_50k))



media_1k = np.mean(lista_1k)
media_10k = np.mean(lista_10k)
media_20k = np.mean(lista_20k)
media_30k = np.mean(lista_30k)
media_40k = np.mean(lista_40k)
media_50k = np.mean(lista_50k)

lista_media = []
lista_media.append(media_1k)
lista_media.append(media_10k)
lista_media.append(media_20k)
lista_media.append(media_30k)
lista_media.append(media_40k)
lista_media.append(media_50k)


#----------------------MUDAR ESSE METODO DE BUSCAR O WORSTCASE-----------------------
arr_wc_1k = np.arange(1000, 0, -1)
arr_wc_10k = np.arange(10000, 0, -1)
arr_wc_20k = np.arange(20000, 0, -1)
arr_wc_30k = np.arange(30000, 0, -1)
arr_wc_40k = np.arange(40000, 0, -1)
arr_wc_50k = np.arange(50000, 0, -1)


pior_caso_tempos = []

pior_caso_tempos.append(calcular_tempo(arr_wc_1k))
pior_caso_tempos.append(calcular_tempo(arr_wc_10k))
pior_caso_tempos.append(calcular_tempo(arr_wc_20k))
pior_caso_tempos.append(calcular_tempo(arr_wc_30k))
pior_caso_tempos.append(calcular_tempo(arr_wc_40k))
pior_caso_tempos.append(calcular_tempo(arr_wc_50k))

lista_a = [lista_1k[0], lista_10k[0], lista_20k[0],lista_30k[0], lista_40k[0], lista_50k[0]]
lista_b = [lista_1k[1], lista_10k[1], lista_20k[1],lista_30k[1], lista_40k[1], lista_50k[1]]
lista_c = [lista_1k[2], lista_10k[2], lista_20k[2],lista_30k[2], lista_40k[2], lista_50k[2]]


fig , ax = plt.subplots(figsize=(10, 6), layout = 'constrained')
ax.plot(tamanhos,lista_media,
    label='Médias dos tempos',
    color='blue',
    linewidth = 5,
    linestyle = ':')

ax.plot(tamanhos, pior_caso_tempos,
    label='Piores tempos possíveis',
    color='red',
    linestyle = '--',
    linewidth = 2)

ax.plot(tamanhos, lista_a,
    label='Lista A',
    color='green',
    linewidth = 1)

ax.plot(tamanhos, lista_b,
    label='Lista B',
    color='yellow',
    linewidth = 1)

ax.plot(tamanhos, lista_c,
    label='Lista C',
    color='orange',
    linewidth = 1)

ax.set_xlabel('Tamanho da Lista')
ax.set_ylabel('Tempo (segundos)')
ax.set_title('Tempo de Ordenação vs Tamanho da Lista')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

# SHELL SORT


```python
import random
import matplotlib.pyplot as plt
import time
import numpy as np

def shell_sort(lista):
  gap = len(lista)//2
  while gap > 0:
    for i in range(gap,len(lista)):
        aux = lista[i]
        j = i
        while j >= gap and lista[j - gap] > aux :
            lista[j] = lista[j - gap]
            j = j - gap
        lista[j] = aux

    gap = gap//2

  return lista

def calcular_tempo(lista_ordenada):
  start_time = time.time()
  shell_sort(lista_ordenada)
  end_time = time.time()
  total_time = end_time - start_time
  return total_time



# Listas para armazenar os tempos
tempos_lista1k = []
tempos_lista10k = []
tempos_lista20k = []
tempo_lista30k = []
tempo_lista40k = []
tempo_lista50k = []
tempos_media = []
tempos_pior_caso = []


tamanhos = [1000, 10000 , 20000, 30000, 40000, 50000]
lista_a_1k = np.random.randint(1, 101, 1000)
lista_b_1k = np.random.randint(1, 101, 1000)
lista_c_1k = np.random.randint(1, 101, 1000)
lista_a_10k = np.random.randint(1, 1001, 10000)
lista_b_10k = np.random.randint(1, 1001, 10000)
lista_c_10k = np.random.randint(1, 1001, 10000)
lista_a_20k = np.random.randint(1, 2001, 20000)
lista_b_20k = np.random.randint(1, 2001, 20000)
lista_c_20k  = np.random.randint(1, 2001, 20000)
lista_a_30k = np.random.randint(1, 3001, 30000)
lista_b_30k = np.random.randint(1, 3001, 30000)
lista_c_30k = np.random.randint(1, 3001, 30000)
lista_a_40k = np.random.randint(1, 4001, 40000)
lista_b_40k = np.random.randint(1, 4001, 40000)
lista_c_40k = np.random.randint(1, 4001, 40000)
lista_a_50k = np.random.randint(1, 5001, 50000)
lista_b_50k = np.random.randint(1, 5001, 50000)
lista_c_50k = np.random.randint(1, 5001, 50000)

def tempo_listas(lista_a, lista_b, lista_c):
  t1 = calcular_tempo(lista_a)
  t2 = calcular_tempo(lista_b)
  t3 = calcular_tempo(lista_c)
  return t1, t2, t3

lista_1k  = list(tempo_listas(lista_a_1k, lista_b_1k, lista_c_1k))
lista_10k = list(tempo_listas(lista_a_10k, lista_b_10k, lista_c_10k))
lista_20k = list(tempo_listas(lista_a_20k, lista_b_20k, lista_c_20k))
lista_30k = list(tempo_listas(lista_a_30k, lista_b_30k, lista_c_30k))
lista_40k = list(tempo_listas(lista_a_40k, lista_b_40k, lista_c_40k))
lista_50k = list(tempo_listas(lista_a_50k, lista_b_50k, lista_c_50k))



media_1k = np.mean(lista_1k)
media_10k = np.mean(lista_10k)
media_20k = np.mean(lista_20k)
media_30k = np.mean(lista_30k)
media_40k = np.mean(lista_40k)
media_50k = np.mean(lista_50k)

lista_media = []
lista_media.append(media_1k)
lista_media.append(media_10k)
lista_media.append(media_20k)
lista_media.append(media_30k)
lista_media.append(media_40k)
lista_media.append(media_50k)


#----------------------MUDAR ESSE METODO DE BUSCAR O WORSTCASE-----------------------
arr_wc_1k = np.arange(1000, 0, -1)
arr_wc_10k = np.arange(10000, 0, -1)
arr_wc_20k = np.arange(20000, 0, -1)
arr_wc_30k = np.arange(30000, 0, -1)
arr_wc_40k = np.arange(40000, 0, -1)
arr_wc_50k = np.arange(50000, 0, -1)


pior_caso_tempos = []

pior_caso_tempos.append(calcular_tempo(arr_wc_1k))
pior_caso_tempos.append(calcular_tempo(arr_wc_10k))
pior_caso_tempos.append(calcular_tempo(arr_wc_20k))
pior_caso_tempos.append(calcular_tempo(arr_wc_30k))
pior_caso_tempos.append(calcular_tempo(arr_wc_40k))
pior_caso_tempos.append(calcular_tempo(arr_wc_50k))

lista_a = [lista_1k[0], lista_10k[0], lista_20k[0],lista_30k[0], lista_40k[0], lista_50k[0]]
lista_b = [lista_1k[1], lista_10k[1], lista_20k[1],lista_30k[1], lista_40k[1], lista_50k[1]]
lista_c = [lista_1k[2], lista_10k[2], lista_20k[2],lista_30k[2], lista_40k[2], lista_50k[2]]


fig , ax = plt.subplots(figsize=(10, 6), layout = 'constrained')
ax.plot(tamanhos,lista_media,
    label='Médias dos tempos',
    color='blue',
    linewidth = 5,
    linestyle = ':')

ax.plot(tamanhos, pior_caso_tempos,
    label='Piores tempos possíveis',
    color='red',
    linestyle = '--',
    linewidth = 2)

ax.plot(tamanhos, lista_a,
    label='Lista A',
    color='green',
    linewidth = 1)

ax.plot(tamanhos, lista_b,
    label='Lista B',
    color='yellow',
    linewidth = 1)

ax.plot(tamanhos, lista_c,
    label='Lista C',
    color='orange',
    linewidth = 1)

ax.set_xlabel('Tamanho da Lista')
ax.set_ylabel('Tempo (segundos)')
ax.set_title('Tempo de Ordenação vs Tamanho da Lista')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

# Quick Sort

```python
import random
import matplotlib.pyplot as plt
import time
import numpy as np

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quick_sort(left) + middle + quick_sort(right)

def calcular_tempo(lista_ordenada):
  start_time = time.time()
  quick_sort(lista_ordenada)
  end_time = time.time()
  total_time = end_time - start_time
  return total_time


# Listas para armazenar os tempos
tempos_lista1k = []
tempos_lista10k = []
tempos_lista20k = []
tempo_lista30k = []
tempo_lista40k = []
tempo_lista50k = []
tempos_media = []
tempos_pior_caso = []

tamanhos = [1000, 10000 , 20000, 30000, 40000, 50000]
lista_a_1k = np.random.randint(1, 101, 1000)
lista_b_1k = np.random.randint(1, 101, 1000)
lista_c_1k = np.random.randint(1, 101, 1000)
lista_a_10k = np.random.randint(1, 1001, 10000)
lista_b_10k = np.random.randint(1, 1001, 10000)
lista_c_10k = np.random.randint(1, 1001, 10000)
lista_a_20k = np.random.randint(1, 2001, 20000)
lista_b_20k = np.random.randint(1, 2001, 20000)
lista_c_20k  = np.random.randint(1, 2001, 20000)
lista_a_30k = np.random.randint(1, 3001, 30000)
lista_b_30k = np.random.randint(1, 3001, 30000)
lista_c_30k = np.random.randint(1, 3001, 30000)
lista_a_40k = np.random.randint(1, 4001, 40000)
lista_b_40k = np.random.randint(1, 4001, 40000)
lista_c_40k = np.random.randint(1, 4001, 40000)
lista_a_50k = np.random.randint(1, 5001, 50000)
lista_b_50k = np.random.randint(1, 5001, 50000)
lista_c_50k = np.random.randint(1, 5001, 50000)

def tempo_listas(lista_a, lista_b, lista_c):
  t1 = calcular_tempo(lista_a)
  t2 = calcular_tempo(lista_b)
  t3 = calcular_tempo(lista_c)
  return t1, t2, t3

lista_1k  = list(tempo_listas(lista_a_1k, lista_b_1k, lista_c_1k))
lista_10k = list(tempo_listas(lista_a_10k, lista_b_10k, lista_c_10k))
lista_20k = list(tempo_listas(lista_a_20k, lista_b_20k, lista_c_20k))
lista_30k = list(tempo_listas(lista_a_30k, lista_b_30k, lista_c_30k))
lista_40k = list(tempo_listas(lista_a_40k, lista_b_40k, lista_c_40k))
lista_50k = list(tempo_listas(lista_a_50k, lista_b_50k, lista_c_50k))



media_1k = np.mean(lista_1k)
media_10k = np.mean(lista_10k)
media_20k = np.mean(lista_20k)
media_30k = np.mean(lista_30k)
media_40k = np.mean(lista_40k)
media_50k = np.mean(lista_50k)

lista_media = []
lista_media.append(media_1k)
lista_media.append(media_10k)
lista_media.append(media_20k)
lista_media.append(media_30k)
lista_media.append(media_40k)
lista_media.append(media_50k)


#----------------------MUDAR ESSE METODO DE BUSCAR O WORSTCASE-----------------------
arr_wc_1k = np.arange(1000, 0, -1)
arr_wc_10k = np.arange(10000, 0, -1)
arr_wc_20k = np.arange(20000, 0, -1)
arr_wc_30k = np.arange(30000, 0, -1)
arr_wc_40k = np.arange(40000, 0, -1)
arr_wc_50k = np.arange(50000, 0, -1)


pior_caso_tempos = []

pior_caso_tempos.append(calcular_tempo(arr_wc_1k))
pior_caso_tempos.append(calcular_tempo(arr_wc_10k))
pior_caso_tempos.append(calcular_tempo(arr_wc_20k))
pior_caso_tempos.append(calcular_tempo(arr_wc_30k))
pior_caso_tempos.append(calcular_tempo(arr_wc_40k))
pior_caso_tempos.append(calcular_tempo(arr_wc_50k))

lista_a = [lista_1k[0], lista_10k[0], lista_20k[0],lista_30k[0], lista_40k[0], lista_50k[0]]
lista_b = [lista_1k[1], lista_10k[1], lista_20k[1],lista_30k[1], lista_40k[1], lista_50k[1]]
lista_c = [lista_1k[2], lista_10k[2], lista_20k[2],lista_30k[2], lista_40k[2], lista_50k[2]]


fig , ax = plt.subplots(figsize=(10, 6), layout = 'constrained')
ax.plot(tamanhos,lista_media,
    label='Médias dos tempos',
    color='blue',
    linewidth = 5,
    linestyle = ':')

ax.plot(tamanhos, pior_caso_tempos,
    label='Piores tempos possíveis',
    color='red',
    linestyle = '--',
    linewidth = 2)

ax.plot(tamanhos, lista_a,
    label='Lista A',
    color='green',
    linewidth = 1)

ax.plot(tamanhos, lista_b,
    label='Lista B',
    color='yellow',
    linewidth = 1)

ax.plot(tamanhos, lista_c,
    label='Lista C',
    color='orange',
    linewidth = 1)

ax.set_xlabel('Tamanho da Lista')
ax.set_ylabel('Tempo (segundos)')
ax.set_title('Tempo de Ordenação vs Tamanho da Lista')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

# Bucket *Sort*

```python
import random
import matplotlib.pyplot as plt
import time
import numpy as np

def bucket_sort(arr):
    if not arr:
        return []

    max_val = max(arr)
    min_val = min(arr)


    if max_val == min_val:
        return arr[:]

    num_buckets = 10
    buckets = [[] for _ in range(num_buckets)]

    range_of_values = max_val - min_val + 1

    for num in arr:
        bucket_index = int((num - min_val) * num_buckets / range_of_values)
        if bucket_index >= num_buckets:
            bucket_index = num_buckets - 1
        buckets[bucket_index].append(num)

    sorted_arr = []
    for bucket in buckets:
        bucket.sort() # Sort individual buckets
        sorted_arr.extend(bucket)

    return sorted_arr


def calcular_tempo(lista_ordenada):
  start_time = time.time()
  bucket_sort(lista_ordenada)
  end_time = time.time()
  total_time = end_time - start_time
  return total_time


# Listas para armazenar os tempos
tempos_lista1k = []
tempos_lista10k = []
tempos_lista20k = []
tempo_lista30k = []
tempo_lista40k = []
tempo_lista50k = []
tempos_media = []
tempos_pior_caso = []


tamanhos = [1000, 10000 , 20000, 30000, 40000, 50000]
lista_a_1k = np.random.randint(1, 101, 1000)
lista_b_1k = np.random.randint(1, 101, 1000)
lista_c_1k = np.random.randint(1, 101, 1000)
lista_a_10k = np.random.randint(1, 1001, 10000)
lista_b_10k = np.random.randint(1, 1001, 10000)
lista_c_10k = np.random.randint(1, 1001, 10000)
lista_a_20k = np.random.randint(1, 2001, 20000)
lista_b_20k = np.random.randint(1, 2001, 20000)
lista_c_20k  = np.random.randint(1, 2001, 20000)
lista_a_30k = np.random.randint(1, 3001, 30000)
lista_b_30k = np.random.randint(1, 3001, 30000)
lista_c_30k = np.random.randint(1, 3001, 30000)
lista_a_40k = np.random.randint(1, 4001, 40000)
lista_b_40k = np.random.randint(1, 4001, 40000)
lista_c_40k = np.random.randint(1, 4001, 40000)
lista_a_50k = np.random.randint(1, 5001, 50000)
lista_b_50k = np.random.randint(1, 5001, 50000)
lista_c_50k = np.random.randint(1, 5001, 50000)

def tempo_listas(lista_a, lista_b, lista_c):
  t1 = calcular_tempo(list(lista_a))
  t2 = calcular_tempo(list(lista_b))
  t3 = calcular_tempo(list(lista_c))
  return t1, t2, t3

lista_1k  = list(tempo_listas(lista_a_1k, lista_b_1k, lista_c_1k))
lista_10k = list(tempo_listas(lista_a_10k, lista_b_10k, lista_c_10k))
lista_20k = list(tempo_listas(lista_a_20k, lista_b_20k, lista_c_20k))
lista_30k = list(tempo_listas(lista_a_30k, lista_b_30k, lista_c_30k))
lista_40k = list(tempo_listas(lista_a_40k, lista_b_40k, lista_c_40k))
lista_50k = list(tempo_listas(lista_a_50k, lista_b_50k, lista_c_50k))



media_1k = np.mean(lista_1k)
media_10k = np.mean(lista_10k)
media_20k = np.mean(lista_20k)
media_30k = np.mean(lista_30k)
media_40k = np.mean(lista_40k)
media_50k = np.mean(lista_50k)

lista_media = []
lista_media.append(media_1k)
lista_media.append(media_10k)
lista_media.append(media_20k)
lista_media.append(media_30k)
lista_media.append(media_40k)
lista_media.append(media_50k)


#----------------------MUDAR ESSE METODO DE BUSCAR O WORSTCASE-----------------------
arr_wc_1k = np.arange(1000, 0, -1)
arr_wc_10k = np.arange(10000, 0, -1)
arr_wc_20k = np.arange(20000, 0, -1)
arr_wc_30k = np.arange(30000, 0, -1)
arr_wc_40k = np.arange(40000, 0, -1)
arr_wc_50k = np.arange(50000, 0, -1)


pior_caso_tempos = []

pior_caso_tempos.append(calcular_tempo(list(arr_wc_1k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_10k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_20k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_30k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_40k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_50k)))

lista_a = [lista_1k[0], lista_10k[0], lista_20k[0],lista_30k[0], lista_40k[0], lista_50k[0]]
lista_b = [lista_1k[1], lista_10k[1], lista_20k[1],lista_30k[1], lista_40k[1], lista_50k[1]]
lista_c = [lista_1k[2], lista_10k[2], lista_20k[2],lista_30k[2], lista_40k[2], lista_50k[2]]


fig , ax = plt.subplots(figsize=(10, 6), layout = 'constrained')
ax.plot(tamanhos,lista_media,
    label='Médias dos tempos',
    color='blue',
    linewidth = 5,
    linestyle = ':')

ax.plot(tamanhos, pior_caso_tempos,
    label='Piores tempos possíveis',
    color='red',
    linestyle = '--',
    linewidth = 2)

ax.plot(tamanhos, lista_a,
    label='Lista A',
    color='green',
    linewidth = 1)

ax.plot(tamanhos, lista_b,
    label='Lista B',
    color='yellow',
    linewidth = 1)

ax.plot(tamanhos, lista_c,
    label='Lista C',
    color='orange',
    linewidth = 1)

ax.set_xlabel('Tamanho da Lista')
ax.set_ylabel('Tempo (segundos)')
ax.set_title('Tempo de Ordenação vs Tamanho da Lista')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

# Radix Sort

```python
import random
import matplotlib.pyplot as plt
import time
import numpy as np

def radix_sort(arr):
    max_num = max(arr)
    exp = 1

    while max_num // exp > 0:
        n = len(arr)
        output = [0] * n
        count = [0] * 10
        for num in arr:
            index = (num // exp) % 10
            count[index] += 1
        for i in range(1, 10):
            count[i] += count[i - 1]
        for i in range(n - 1, -1, -1):
            index = (arr[i] // exp) % 10
            output[count[index] - 1] = arr[i]
            count[index] -= 1
        for i in range(n):
            arr[i] = output[i]

        exp *= 10


def calcular_tempo(lista_ordenada):
  start_time = time.time()
  radix_sort(lista_ordenada)
  end_time = time.time()
  total_time = end_time - start_time
  return total_time


# Listas para armazenar os tempos
tempos_lista1k = []
tempos_lista10k = []
tempos_lista20k = []
tempo_lista30k = []
tempo_lista40k = []
tempo_lista50k = []
tempos_media = []
tempos_pior_caso = []


tamanhos = [1000, 10000 , 20000, 30000, 40000, 50000]
lista_a_1k = np.random.randint(1, 101, 1000)
lista_b_1k = np.random.randint(1, 101, 1000)
lista_c_1k = np.random.randint(1, 101, 1000)
lista_a_10k = np.random.randint(1, 1001, 10000)
lista_b_10k = np.random.randint(1, 1001, 10000)
lista_c_10k = np.random.randint(1, 1001, 10000)
lista_a_20k = np.random.randint(1, 2001, 20000)
lista_b_20k = np.random.randint(1, 2001, 20000)
lista_c_20k  = np.random.randint(1, 2001, 20000)
lista_a_30k = np.random.randint(1, 3001, 30000)
lista_b_30k = np.random.randint(1, 3001, 30000)
lista_c_30k = np.random.randint(1, 3001, 30000)
lista_a_40k = np.random.randint(1, 4001, 40000)
lista_b_40k = np.random.randint(1, 4001, 40000)
lista_c_40k = np.random.randint(1, 4001, 40000)
lista_a_50k = np.random.randint(1, 5001, 50000)
lista_b_50k = np.random.randint(1, 5001, 50000)
lista_c_50k = np.random.randint(1, 5001, 50000)

def tempo_listas(lista_a, lista_b, lista_c):
  t1 = calcular_tempo(list(lista_a))
  t2 = calcular_tempo(list(lista_b))
  t3 = calcular_tempo(list(lista_c))
  return t1, t2, t3

lista_1k  = list(tempo_listas(lista_a_1k, lista_b_1k, lista_c_1k))
lista_10k = list(tempo_listas(lista_a_10k, lista_b_10k, lista_c_10k))
lista_20k = list(tempo_listas(lista_a_20k, lista_b_20k, lista_c_20k))
lista_30k = list(tempo_listas(lista_a_30k, lista_b_30k, lista_c_30k))
lista_40k = list(tempo_listas(lista_a_40k, lista_b_40k, lista_c_40k))
lista_50k = list(tempo_listas(lista_a_50k, lista_b_50k, lista_c_50k))



media_1k = np.mean(lista_1k)
media_10k = np.mean(lista_10k)
media_20k = np.mean(lista_20k)
media_30k = np.mean(lista_30k)
media_40k = np.mean(lista_40k)
media_50k = np.mean(lista_50k)

lista_media = []
lista_media.append(media_1k)
lista_media.append(media_10k)
lista_media.append(media_20k)
lista_media.append(media_30k)
lista_media.append(media_40k)
lista_media.append(media_50k)


#----------------------MUDAR ESSE METODO DE BUSCAR O WORSTCASE-----------------------
arr_wc_1k = np.arange(1000, 0, -1)
arr_wc_10k = np.arange(10000, 0, -1)
arr_wc_20k = np.arange(20000, 0, -1)
arr_wc_30k = np.arange(30000, 0, -1)
arr_wc_40k = np.arange(40000, 0, -1)
arr_wc_50k = np.arange(50000, 0, -1)


pior_caso_tempos = []

pior_caso_tempos.append(calcular_tempo(list(arr_wc_1k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_10k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_20k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_30k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_40k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_50k)))

lista_a = [lista_1k[0], lista_10k[0], lista_20k[0],lista_30k[0], lista_40k[0], lista_50k[0]]
lista_b = [lista_1k[1], lista_10k[1], lista_20k[1],lista_30k[1], lista_40k[1], lista_50k[1]]
lista_c = [lista_1k[2], lista_10k[2], lista_20k[2],lista_30k[2], lista_40k[2], lista_50k[2]]


fig , ax = plt.subplots(figsize=(10, 6), layout = 'constrained')
ax.plot(tamanhos,lista_media,
    label='Médias dos tempos',
    color='blue',
    linewidth = 5,
    linestyle = ':')

ax.plot(tamanhos, pior_caso_tempos,
    label='Piores tempos possíveis',
    color='red',
    linestyle = '--',
    linewidth = 2)

ax.plot(tamanhos, lista_a,
    label='Lista A',
    color='green',
    linewidth = 1)

ax.plot(tamanhos, lista_b,
    label='Lista B',
    color='yellow',
    linewidth = 1)

ax.plot(tamanhos, lista_c,
    label='Lista C',
    color='orange',
    linewidth = 1)

ax.set_xlabel('Tamanho da Lista')
ax.set_ylabel('Tempo (segundos)')
ax.set_title('Tempo de Ordenação vs Tamanho da Lista')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

# Heap Sort

```python
def heapify(arr, n, i):
    largest = i
    left_child = 2 * i + 1
    right_child = 2 * i + 2

    if left_child < n and arr[left_child] > arr[largest]:
        largest = left_child

    if right_child < n and arr[right_child] > arr[largest]:
        largest = right_child
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):

    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]

        heapify(arr, i, 0)

def calcular_tempo(lista_ordenada):
  start_time = time.time()
  heap_sort(lista_ordenada)
  end_time = time.time()
  total_time = end_time - start_time
  return total_time


# Listas para armazenar os tempos
tempos_lista1k = []
tempos_lista10k = []
tempos_lista20k = []
tempo_lista30k = []
tempo_lista40k = []
tempo_lista50k = []
tempos_media = []
tempos_pior_caso = []


tamanhos = [1000, 10000 , 20000, 30000, 40000, 50000]
lista_a_1k = np.random.randint(1, 101, 1000)
lista_b_1k = np.random.randint(1, 101, 1000)
lista_c_1k = np.random.randint(1, 101, 1000)
lista_a_10k = np.random.randint(1, 1001, 10000)
lista_b_10k = np.random.randint(1, 1001, 10000)
lista_c_10k = np.random.randint(1, 1001, 10000)
lista_a_20k = np.random.randint(1, 2001, 20000)
lista_b_20k = np.random.randint(1, 2001, 20000)
lista_c_20k  = np.random.randint(1, 2001, 20000)
lista_a_30k = np.random.randint(1, 3001, 30000)
lista_b_30k = np.random.randint(1, 3001, 30000)
lista_c_30k = np.random.randint(1, 3001, 30000)
lista_a_40k = np.random.randint(1, 4001, 40000)
lista_b_40k = np.random.randint(1, 4001, 40000)
lista_c_40k = np.random.randint(1, 4001, 40000)
lista_a_50k = np.random.randint(1, 5001, 50000)
lista_b_50k = np.random.randint(1, 5001, 50000)
lista_c_50k = np.random.randint(1, 5001, 50000)

def tempo_listas(lista_a, lista_b, lista_c):
  t1 = calcular_tempo(list(lista_a))
  t2 = calcular_tempo(list(lista_b))
  t3 = calcular_tempo(list(lista_c))
  return t1, t2, t3

lista_1k  = list(tempo_listas(lista_a_1k, lista_b_1k, lista_c_1k))
lista_10k = list(tempo_listas(lista_a_10k, lista_b_10k, lista_c_10k))
lista_20k = list(tempo_listas(lista_a_20k, lista_b_20k, lista_c_20k))
lista_30k = list(tempo_listas(lista_a_30k, lista_b_30k, lista_c_30k))
lista_40k = list(tempo_listas(lista_a_40k, lista_b_40k, lista_c_40k))
lista_50k = list(tempo_listas(lista_a_50k, lista_b_50k, lista_c_50k))



media_1k = np.mean(lista_1k)
media_10k = np.mean(lista_10k)
media_20k = np.mean(lista_20k)
media_30k = np.mean(lista_30k)
media_40k = np.mean(lista_40k)
media_50k = np.mean(lista_50k)

lista_media = []
lista_media.append(media_1k)
lista_media.append(media_10k)
lista_media.append(media_20k)
lista_media.append(media_30k)
lista_media.append(media_40k)
lista_media.append(media_50k)


#----------------------MUDAR ESSE METODO DE BUSCAR O WORSTCASE-----------------------
arr_wc_1k = np.arange(1000, 0, -1)
arr_wc_10k = np.arange(10000, 0, -1)
arr_wc_20k = np.arange(20000, 0, -1)
arr_wc_30k = np.arange(30000, 0, -1)
arr_wc_40k = np.arange(40000, 0, -1)
arr_wc_50k = np.arange(50000, 0, -1)


pior_caso_tempos = []

pior_caso_tempos.append(calcular_tempo(list(arr_wc_1k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_10k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_20k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_30k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_40k)))
pior_caso_tempos.append(calcular_tempo(list(arr_wc_50k)))

lista_a = [lista_1k[0], lista_10k[0], lista_20k[0],lista_30k[0], lista_40k[0], lista_50k[0]]
lista_b = [lista_1k[1], lista_10k[1], lista_20k[1],lista_30k[1], lista_40k[1], lista_50k[1]]
lista_c = [lista_1k[2], lista_10k[2], lista_20k[2],lista_30k[2], lista_40k[2], lista_50k[2]]


fig , ax = plt.subplots(figsize=(10, 6), layout = 'constrained')
ax.plot(tamanhos,lista_media,
    label='Médias dos tempos',
    color='blue',
    linewidth = 5,
    linestyle = ':')

ax.plot(tamanhos, pior_caso_tempos,
    label='Piores tempos possíveis',
    color='red',
    linestyle = '--',
    linewidth = 2)

ax.plot(tamanhos, lista_a,
    label='Lista A',
    color='green',
    linewidth = 1)

ax.plot(tamanhos, lista_b,
    label='Lista B',
    color='yellow',
    linewidth = 1)

ax.plot(tamanhos, lista_c,
    label='Lista C',
    color='orange',
    linewidth = 1)

ax.set_xlabel('Tamanho da Lista')
ax.set_ylabel('Tempo (segundos)')
ax.set_title('Tempo de Ordenação vs Tamanho da Lista')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```
